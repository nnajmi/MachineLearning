## Data Preprocessing

## Model, Error and Plot Functions

Below we have three major functions that are going to help us apply the model to our time series data:
<br>
<br>**errorCalculation** function calculates the *Mean Absolute Error* and the *Mean Squared Error* for all AQIs.
We can evaluate our model based on the amount of these errors.
<br>
<br>**plotData** plots the train, test and predicted values of AQIs in darkgray, blue and red respectively.
Note that although the train data contains more records, we decide to depict the lat two years of it.
<br>
<br>**ARModel** applys AR model to our train data using the input parameters and use the generated model to predict the AQI for the test set.

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

pollutants = ['NO2 AQI', 'O3 AQI', 'SO2 AQI', 'CO AQI']
MSEdf = pd.DataFrame(index=['ARFrom2000', 'ARMonthlyFrom2000', 'ARFrom2005', 'ARFrom2010'])
MAEdf = pd.DataFrame(index=['ARFrom2000', 'ARMonthlyFrom2000', 'ARFrom2005', 'ARFrom2010'])

def errorCalculation(method, predict, train, test):
    NONAN_predict = pd.DataFrame()
    NONAN_test = pd.DataFrame()
    for pol in pollutants:
        index = []
        index = predict[pol].index[predict[pol].apply(np.isnan)]
        NONAN_predict[pol] = predict[pol].drop(index)
        NONAN_test[pol] = test[pol].drop(index)
        
        MAE = mean_absolute_error(NONAN_test[pol], NONAN_predict[pol])
        MSE = mean_squared_error(NONAN_test[pol], NONAN_predict[pol])
        MSEdf.loc[method,pol] = np.sqrt(MSE)
        MAEdf.loc[method,pol] = np.sqrt(MAE)
    return MSEdf.loc[method], MAEdf.loc[method]
```

```python
def plotData(predict):
    fig, axes = plt.subplots(2, 2, figsize=(16,7), dpi=100)
    fig.subplots_adjust(wspace=0.1, hspace=0.3)
    axlist = [(0, 0), (0, 1), (1, 0), (1, 1)]

    for (i, j), pol in zip(axlist, pollutants):
        axes[i][j].plot(train[train.index>='2013'].index, train.loc['2013':, pol], lw = 0.75, color = 'darkgray', label = 'Train')
        axes[i][j].plot(test.index, test[pol], lw = 0.75, color = 'blue', label = 'Test', )
        axes[i][j].plot(test.index, predict[pol], c = 'red', lw = 0.75, label = 'Forecast')
        axes[i][j].set_title(pol)
        axes[i][j].legend(ncol=3)
```

## AutoRegressive Model

Autoregressive models predict series only based on the past values in the series that are called *lags*.
<br>These models are called *long memory* models. It means that the predicted value depends on all the previous values however, the dependence of previous observations declines over time.
<br>
<br>The lags parameter in AutoReg model can be an integer or a list. 
<br>lags = 5 will include lags 1 to 5, while lags = [1,5] will only include lags 1 and 5.

```python
# AR
from statsmodels.tsa.ar_model import AutoReg

def ARModel(train, test, params):
    ARpredict = pd.DataFrame()
    # fit model
    NO2_AQI_model = AutoReg(train["NO2 AQI"], lags=params).fit()
    O3_AQI_model = AutoReg(train["O3 AQI"], lags=params).fit()
    SO2_AQI_model = AutoReg(train["SO2 AQI"], lags=params).fit()
    CO_AQI_model = AutoReg(train["CO AQI"], lags=params).fit()

    # make prediction
    ARpredict["NO2 AQI"] = NO2_AQI_model.predict(end=test.shape[0]-1)
    ARpredict["O3 AQI"] = O3_AQI_model.predict(end=test.shape[0]-1)
    ARpredict["SO2 AQI"] = SO2_AQI_model.predict(end=test.shape[0]-1)
    ARpredict["CO AQI"] = CO_AQI_model.predict(end=test.shape[0]-1)

    ARpredict.index = test.index

    #print(NO2_AQI_model.summary())
    return ARpredict
```

We applied the model to various train set using various lags.
<br>The first train set consists of all the records from 2000 to 2015. (**train**)
<br>The second train set consists of all the records from 2005 to 2015. (**trainFrom2005**)
<br>The third train set consists of all the records from 2010 to 2015. (**trainFrom2010**)
<br>For all the cases the test set consists of all the records from 2015 to 2016. (**test**) 
We predict the AQI values for the test dates and compare it to the actual value from dataset.
<br>
<br>For the *lags* parameter, two different cases were considered: 
<br>1. *lags = 1* which considers only lag 1
<br>2. *lags = 30* which considers all the lags from 1 to 30.

```python
ARpredict = ARModel(train, test, 1)
ARErrors = errorCalculation('ARFrom2000', ARpredict, train, test)
plotData(ARpredict)
```

```python
ARMonthlypredict = ARModel(train, test, 30)
ARErrors = errorCalculation('ARMonthlyFrom2000', ARMonthlypredict, train, test)
plotData(ARMonthlypredict)
```

```python
trainFrom2005 = AQIData['2005-01-01':'2015-01-01']
print("Percentage of train data compare to the whole data: ", len(trainFrom2005)/(len(trainFrom2005)+len(test))*100)
print("Percentage of test data compare to the whole data: ", len(test)/(len(trainFrom2005)+len(test))*100)
```

```python
ARpredict = ARModel(trainFrom2005, test, 1)
ARErrors = errorCalculation('ARFrom2005', ARpredict, trainFrom2005, test)
plotData(ARpredict)
```
