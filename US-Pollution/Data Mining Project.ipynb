#### Extracting infromation for the state of "California"

```python
data = data[['ID', 'Site Num', 'State', 'County', 'City', 'Date Local','NO2 AQI','O3 AQI','SO2 AQI','CO AQI']]
dataCA = data.loc[data['State'] == 'California']
#dataCA = dataCA.dropna(axis='rows') 
dataCA['Date Local'] = pd.to_datetime(dataCA['Date Local'],format='%m/%d/%Y')  # Change date from string to date value
print("Size of data: ", dataCA.shape)

#dataCA = dataCA.groupby(['Site Num','Date Local'], as_index=False).mean()  # Take mean values if there are depulicated entries
dataCA = dataCA.groupby(['Date Local'], as_index=False).mean()
dataCA.head()
```

Here are two different ways of calculating the mean value of the four pollutants per month per year:


In the first method, we group by the year and then month using the 'Date Local' column 
In the second method, we split the 'Date Local' column to three day, month, year columns and add them to the table and then group by the data using the new columns

```python
#First Method
dataCA.groupby([dataCA['Date Local'].dt.year, dataCA['Date Local'].dt.month]).mean().head(50)

#Second Method
dataCA['day'] = dataCA['Date Local'].dt.day
dataCA['month'] = dataCA['Date Local'].dt.month
dataCA['year'] = dataCA['Date Local'].dt.year
dataCA.groupby(['year', 'month']).mean().head(50)
```
